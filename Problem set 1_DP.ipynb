# 6) Turning a given word into pig-Latin using pre-defined formula :

```python
def pig(word):
    vowel=['a','e','i','o','u']
    
    if word[0] in vowel:
        word=word+"way"
    else:
        word=word[1:]+word[0]+"ay"
    return word
strng=input("Enter a word that you want to turn into pig-latin :")
call=pig(strng)
print(call)
```

# 8) Taking inputs of currencies from user and converting that into US dollars using a given file in which predefined currency convertion is stored.  

```python
file=open('currencies.txt','r')
list_currency = file.read().split()
print(list_currency)

def Curconv(currency,amount):
    for i in list_currency:
        if list_currency[i]==currency:
            multiplier=list_currency[i+1]
            output=multiplier*amount
            print(output)

Curconv('EUR',100)
```

# 10) Creating a function called frequencies() that takes a string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text.

```python
def frequencies(strng):
    
    #Inbuild we have a dictionary of stored character.
    char_dict={}
    #characters dictionary so that we can store the count of each character.
    letters='abcdefghijklmnopqrstuvwxyz'
    
    
    for n in letters:
        char_dict[n]=0
    for n in strng:
        if n in char_dict.keys():
            char_dict[n]=char_dict[n]+1
        else :
            char_dict[n]=1

    print(char_dict.values())
frequencies('The quick red fox got bored and went to home')
```
